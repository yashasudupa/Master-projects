```python
#Part f)
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
import tensorflow as tf
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split

# plt.close('all')  # if you like


# load dataset

with open('isolet_crop_train.pkl', 'rb') as f:
    train_data = pkl.load(f)

with open('isolet_crop_test.pkl', 'rb') as f:
    test_data = pkl.load(f)

X_train, y_train = train_data
X_test, y_test = test_data

#print(X_test.shape, y_test.shape)

#One hot encoding
n_samples,n_inputs = np.shape(X_train)
y_train = y_train.reshape((n_samples,1))
onehot_encoder = OneHotEncoder(sparse=False)
y_coded = onehot_encoder.fit_transform(y_train)

n_samples_test, n_inputs = X_test.shape
y_test = y_test.reshape((n_samples_test, 1))
onehot_encoder = OneHotEncoder(sparse=False)
y_coded_test = onehot_encoder.fit_transform(y_test)


# split the data sets, etc.
xtrain, xearl, ytrain, yearl = train_test_split(X_train, y_coded, test_size=0.15)

scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xval = scaler.transform(xearl)
xtest = scaler.transform(X_test)

yval = yearl
ytest = y_coded_test
#5....Training with the best choice of NN
n_classes = y_coded.shape[1]

input_num_units = n_inputs
hidden_num_units = 150
output_num_units = n_classes



# define placeholders
x = tf.placeholder(tf.float64, [None, input_num_units])
y = tf.placeholder(tf.float64, [None, output_num_units])
prob_out = tf.placeholder(tf.float64)

# set remaining variables
epochs = 5
batch_size = 40
learning_rate = 0.001
d1 = np.sqrt(input_num_units)
d2 = np.sqrt(hidden_num_units)
### define weights and biases of the neural network (refer this article if you don't understand the terminologies)


weights = {
    
    'hidden': tf.Variable(np.random.uniform(low = - (1/d1), high = +(1/d1),size = (input_num_units, hidden_num_units))),
    'output': tf.Variable(np.random.uniform(low = - (1/d2), high = +(1/d2),size = (hidden_num_units, output_num_units)))                     
}

biases = {
    
    'hidden': tf.Variable(np.zeros(hidden_num_units), trainable = True),
    'output': tf.Variable(np.zeros(output_num_units), trainable = True)
}    
    
change_w_hidden = tf.assign(weights['hidden'], np.random.uniform(low = - (1/d1), high = +(1/d1),size = (input_num_units, hidden_num_units)))
change_w_output = tf.assign(weights['output'], np.random.uniform(low = - (1/d1), high = +(1/d1),size = (hidden_num_units, output_num_units))) 
    
hidden_layer = tf.add(tf.matmul(x, weights['hidden']), biases['hidden'])
hidden_layer = tf.nn.relu(hidden_layer)
#Defining dropout
drop_out = tf.nn.dropout(hidden_layer, prob_out)
output_layer =  tf.nn.softmax(tf.matmul(hidden_layer, weights['output']) + biases['output'])


ce = tf.nn.softmax_cross_entropy_with_logits(labels = y, logits = output_layer)
cost = tf.reduce_mean(ce)
correct = tf.equal(tf.argmax(output_layer,1), tf.argmax(y,1))
misclassification = (1 - tf.reduce_mean(tf.cast(correct, tf.float32)))*100

optimizer = tf.train.RMSPropOptimizer(learning_rate=learning_rate).minimize(cost)

init = tf.global_variables_initializer() 
sess = tf.Session()
sess.run(init)

train_error_list = []
test_error_list = []
ear_stop_list = []
estop_mcr_list = []
test_mcr_list = []
train_mcr_list = []
test_error_sum = 0
X_batch_list = np.array_split(xtrain, batch_size)
labels_batch_list = np.array_split(ytrain, batch_size)

for i in range(500):
    for x_minibatch, labels_minibatch in zip(X_batch_list, labels_batch_list):
        sess.run(optimizer, feed_dict={x: x_minibatch, y: labels_minibatch, prob_out: 0.75})
    #implementation of drop out
    train_err = sess.run(cost, feed_dict={x: xtrain, y: ytrain, prob_out: 0.75})
    error_earl = sess.run(cost, feed_dict={x: xval, y: yval, prob_out: 1})
    test_err = sess.run(cost, feed_dict = {x: xtest, y: ytest, prob_out: 1})
    test_error_list.append(test_err)
    ear_stop_list.append(error_earl)
    train_mcr = sess.run(misclassification, feed_dict={x: xtrain, y: ytrain, prob_out: 0.75})
    estop_mcr = sess.run(misclassification, feed_dict={x: xval, y: yval, prob_out: 1})
    test_mcr = sess.run(misclassification, feed_dict ={x: xtest, y: ytest, prob_out:1})
    train_error_list.append(train_err)
    train_mcr_list.append(train_mcr)
    estop_mcr_list.append(estop_mcr)
    test_mcr_list.append(test_mcr)
    if ear_stop_list[i] > ear_stop_list[i-1]:
        break
print("Stopped at:", i)

fig,ax_list = plt.subplots(1,2)
ax_list[0].plot(test_error_list, color='red', label='validation', lw=2)
ax_list[1].plot(train_mcr_list, color='blue', label='training', lw=2)
ax_list[1].plot(test_mcr_list, color='red', label='validation', lw=2)

ax_list[0].set_title('Mean Test Error')
ax_list[0].set_xlabel('Training epoch')
ax_list[0].set_ylabel('Cross-entropy')
ax_list[1].set_title('Misclassification Rate (%)')
ax_list[1].set_xlabel('Training epoch')
ax_list[1].set_ylabel('Misclassification Rate (%)')
ax_list[0].legend(loc='best')
ax_list[1].legend(loc='best')
plt.show()

print("Results over 10 independent runs")
sess.run(change_w_hidden)
sess.run(change_w_output)
ear_stop_list = [] 
test_mcr_sum = 0
for j in range(10):
    for i in range(500):
        for x_minibatch, labels_minibatch in zip(X_batch_list, labels_batch_list):
            sess.run(optimizer, feed_dict={x: x_minibatch, y: labels_minibatch, prob_out:0.75})
        error_earl = sess.run(cost, feed_dict={x: xval, y: yval})
        test_err = sess.run(cost, feed_dict = {x: xtest, y: ytest, prob_out: 1})
        ear_stop_list.append(error_earl)
        test_mcr = sess.run(misclassification, feed_dict ={x: xtest, y: ytest})
        #Implementation of early stopping
        if ear_stop_list[i] > ear_stop_list[i-1]:
            break
    test_mcr_sum = test_mcr_sum + test_mcr   
    ear_stop_list = []
    sess.run(change_w_hidden)
    sess.run(change_w_output)
    print("No. epoch over",j,"th run:",i)  
mean_test_mcr = test_mcr_sum/10
print("Mean % misclassification V:", mean_test_mcr)
sess.close()

```
